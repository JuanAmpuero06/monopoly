### **K-nearest neighbors (KNN) model**

```python
# Select features and target
selected_features = [
    'FlgActCN_T12', 'FlgActCN_T11', 'FlgActCN_T10', 'FlgActCN_T09', 'FlgActCN_T08',
    'FlgActCN_T07', 'FlgActCN_T06', 'FlgActCN_T05', 'FlgActCN_T04', 'FlgActCN_T03',
    'FlgActCN_T02', 'FlgActCN_T01'
]
target = "target"

X = df[selected_features]
y = df[target]

# Apply SMOTE to balance the complete dataset
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)

# Split the data into training (80%) and testing (20%)
X_train, X_test, y_train, y_test = train_test_split(
    X_resampled, y_resampled, test_size=0.20, random_state=42, stratify=y_resampled
)

# Configure and train the KNN model
knn_model = KNeighborsClassifier(
    n_neighbors=5,        # Number of neighbors
    weights="uniform",    # All neighbors have equal weight
    metric="minkowski",   # Default distance metric
    p=2                   # Euclidean distance
)
knn_model.fit(X_train, y_train)

# Evaluate the model on the test set
y_pred_knn = knn_model.predict(X_test)
accuracy = knn_model.score(X_test, y_test)
conf_matrix = confusion_matrix(y_test, y_pred_knn)
class_report = classification_report(y_test, y_pred_knn)

print(f"KNN model accuracy: {accuracy * 100:.2f}%")
print("Confusion Matrix:")
print(conf_matrix)
print("Classification Report:")
print(class_report)

# Cross-validation with details
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
cv_scores = cross_val_score(knn_model, X_resampled, y_resampled, cv=skf, scoring='roc_auc')

print("\n=== AUC-ROC Scores per Fold ===")
for fold_idx, score in enumerate(cv_scores, 1):
    print(f"Fold {fold_idx}: AUC-ROC = {score:.4f}")

print(f"\n=== Average and Standard Deviation ===")
print(f"Average AUC-ROC: {np.mean(cv_scores):.4f} Â± {np.std(cv_scores):.4f}")

# AUC-ROC on the test set
y_test_prob = knn_model.predict_proba(X_test)[:, 1]
roc_auc = roc_auc_score(y_test, y_test_prob)
print(f"\nTest Set AUC-ROC: {roc_auc:.4f}")

# ROC curve
fpr, tpr, thresholds = roc_curve(y_test, y_test_prob)

plt.figure()
plt.plot(fpr, tpr, color="darkorange", lw=2, label=f"ROC curve (area = {roc_auc:.2f})")
plt.plot([0, 1], [0, 1], color="navy", lw=2, linestyle="--")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("Receiver Operating Characteristic (ROC)")
plt.legend(loc="lower right")
plt.show()
```

#### **The 'Tpot' method is used to return which is the most appropriate model based on the established variables.**

```python
# Select features and target variable
selected_features = [
    'FlgActCN_T12', 'FlgActCN_T11', 'FlgActCN_T10', 'FlgActCN_T09', 'FlgActCN_T08',
    'FlgActCN_T07', 'FlgActCN_T06', 'FlgActCN_T05', 'FlgActCN_T04', 'FlgActCN_T03',
    'FlgActCN_T02', 'FlgActCN_T01'
]
target = "target"

X = df[selected_features]
y = df[target]

# Apply SMOTE to balance the dataset
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)

# Split the data into training (80%) and testing (20%) sets
X_train, X_test, y_train, y_test = train_test_split(
    X_resampled, y_resampled, test_size=0.20, random_state=42, stratify=y_resampled
)

# Configure TPOT to find the best model
tpot = TPOTClassifier(
    generations=3,  # Number of generations for pipeline optimization
    population_size=10,  # Population size per generation
    cv=3,  # Number of cross-validation folds
    scoring='roc_auc', # Metric to maximize
    verbosity=2, # Verbosity level for logs
    random_state=42,
    n_jobs=-1 # Use all available CPU cores
)

# Train TPOT
tpot.fit(X_train, y_train)

# Evaluate the model on the test set
y_test_pred = tpot.predict(X_test)
y_test_prob = tpot.predict_proba(X_test)[:, 1]

# Print classification metrics
print("\n=== Classification Report ===")
print(classification_report(y_test, y_test_pred))

print("\n=== Confusion Matrix ===")
print(confusion_matrix(y_test, y_test_pred))

# Calculate and display AUC-ROC
roc_auc = roc_auc_score(y_test, y_test_prob)
print(f"\nTest Set AUC-ROC: {roc_auc:.4f}")

# Plot ROC curve
fpr, tpr, thresholds = roc_curve(y_test, y_test_prob)
plt.figure()
plt.plot(fpr, tpr, color="darkorange", lw=2, label=f"ROC curve (area = {roc_auc:.2f})")
plt.plot([0, 1], [0, 1], color="navy", lw=2, linestyle="--")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("Receiver Operating Characteristic (ROC)")
plt.legend(loc="lower right")
plt.show()

# Export the best pipeline
tpot.export('best_model_pipeline.py')
```
